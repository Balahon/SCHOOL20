```python
a = "1"*104
print(a)
while("111" in a):
# вернет копию строки, в которой все вхождения подстроки old 
# заменены на подстроку new.
# Если указан необязательный аргумент count, 
# заменяются только первые вхождения count.
# old - str, подстрока или символ который меняем
# new - str, подстрока или символ на который меняем
# count - int, сколько раз меняем
    a = a.replace('111',"22",1) 
    a = a.replace("222",'11',1)
    print(a)
print(a)
    
```

```python
strng = '1' * 104
print()
while ('111' in strng):
    strng = strng.replace('111', '22', 1)
    strng = strng.replace('222', '11', 1)
print(strng)
```

```python
a = "1" + "9"*100
print(a)
while(("19" in a) or ("299" in a) or("3999" in a)):
    a = a.replace("19", "2", 1)
    print(a)
    a = a.replace("299", "3", 1)
    print(a)
    a = a.replace("3999", "1", 1)
    print(a)
print("результат : ", a)
```

```python
a = "8"*70
print(a)
while(("2222" in a) or ("8888" in a)):
    a = a.replace("2222", "88", 1)
    print(a)
    a = a.replace("8888", "22", 1)
    print(a)
print("результат : ", a)
```
